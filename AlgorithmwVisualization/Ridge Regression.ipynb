# Ridge Regression is an algorithm used to solve overfitting

The objective function of ridge regression is added with a penalty Σ(w_i)^2 where λ is the regularization parameter. The new objective function becomes:

---
 **J(w, b) = Σ((y - Σ(w_i * x_i) - b)^2) + λ * Σ(w_i)^2**

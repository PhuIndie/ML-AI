```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)
X = np.random.rand(1000,1)
y = 4 + 3*X + .2*np.random.rand(1000,1) # Linear Regression Example
# Khởi tạo Xbar
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis=1)
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w_lr = np.dot(np.linalg.pinv(A), b)
# Mỗi điểm dữ liệu gradient (epoch)
def sgrad(w, i, rd_id) :
    true_i = rd_id[i]
    xi = Xbar[true_i, :]
    yi = y[true_i]
    a = np.dot(xi, w) - yi
    return (xi*a).reshape(2,1)

def grad(w):
    N = Xbar.shape[0]
    return 1/N * Xbar.T.dot(Xbar.dot(w) - y)

# Stochastic Gradient Descent
def SGD(w_init, grad, eta):
    w = [w_init]
    w_last_check = w_init
    iter_check_w = 10
    N = X.shape[0]
    count = 0
    for it in range(1000) :
        rd_id = np.random.permutation(N)
        for i in range(N) :
            count += 1
            g = sgrad(w[-1], i, rd_id)
            w_new = w[-1] - eta*g
            w.append(w_new)
            if count%iter_check_w == 0:
                w_this_check = w_new
                if np.linalg.norm(w_this_check-w_last_check)/len(w_init) < 1e-3 :
                    return (w,it)
                w_last_check = w_this_check
    return (w,it)

w_init1 = np.array([[2], [1]])
w_init2 = np.array([[6], [4]])
(w1, init1) = SGD(w_init1, sgrad, .7)
(w2, init2) = SGD(w_init2, sgrad, .7)
print('Stochastic Gradient Descent for w[2,1] =', w1[-1].T, 'with iterations =', init1+1)
print('Stochastic Gradient Descent for w[6,4] =', w2[-1].T, 'with iterations =', init2+1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
np.random.seed(2)

means = [[2, 2], [4, 2]]
cov = [[.3, .2], [.2, .3]]
N = 10
X0 = np.random.multivariate_normal(means[0], cov, N).T
X1 = np.random.multivariate_normal(means[1], cov, N).T

X = np.concatenate((X0, X1), axis=1)
y = np.concatenate((np.ones((1,N)), -1*np.ones((1,N))), axis=1)

#Xbar
X = np.concatenate((np.ones((1,2*N)), X), axis=0)

# PLA
def h(w, x) :
    return np.sign(np.dot(w.T,x))  # tính đầu ra khi biết đầu vào x là weights w

def has_coverage(X, y, w):
    return np.array_equal(h(w,X), y) # kiểm tra xem thuật toán đã hội tụ chưa

def perceptron(X, y, w_init) :
    w = [w_init]
    N = X.shape[1]
    d = X.shape[0]
    mis_points = []
    while True :
        # mix data
        mix_id = np.random.permutation(N)
        for i in range(N):
            xi = X[:, mix_id[i]].reshape(d,1)
            yi = y[0, mix_id[i]]
            if h(w[-1], xi)[0] != yi:
                mis_points.append(mix_id[i])
                w_new = w[-1] + yi*xi
                w.append(w_new)

        if has_coverage(X, y, w[-1]):
            break
    return (w, mis_points,i)

d = X.shape[0]
w_init = np.random.randn(d,1)
(w,m,i) = perceptron(X, y, w_init)
print('w = ', w[-1].T, 'm =', m, 'iteration =',i-1)
```
